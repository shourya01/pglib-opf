```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
```

```python
data118 = np.load(os.getcwd()+'/data/pglib_opf_case118_ieee_inp.npz')
data2312 = np.load(os.getcwd()+'/data/pglib_opf_case2312_goc_inp.npz')
data4601 = np.load(os.getcwd()+'/data/pglib_opf_case4601_goc_inp.npz')
data10000 = np.load(os.getcwd()+'/data/pglib_opf_case10000_goc_inp.npz')
datas = [data118['data'],data2312['data'],data4601['data'],data10000['data']]
names = ['pglib_opf_case118_ieee','pglib_opf_case2312_goc','pglib_opf_case4601_goc','pglib_opf_case10000_goc']
```

```python
plt.rcParams.update({'font.size':14})
fig, axs = plt.subplots(1,4,figsize=(50//2,10//2))
for idx,(data,name) in enumerate(zip(datas,names)):
    cax = axs[idx].imshow(data==0)
    axs[idx].set_aspect(data.shape[1]/data.shape[0], adjustable='box')
    fig.colorbar(cax, ax=axs[idx])
    axs[idx].set_title(f'{name} inputs\n0=zero, 1=nonzero')
plt.savefig('inputs.png',format='png')
plt.show()
```
